### Accidentes por año

- Se observa una tendencia general al alza desde 2015 hasta 2018, con un aumento constante en el número de accidentes. Este período de crecimiento puede indicar un aumento de la actividad vial, urbanización o cambios en el comportamiento del conductor.
- El año 2018 destaca como el de mayor cantidad de accidentes, con un total de 34,000 incidentes registrados. Este pico merece una atención especial, ya que podría indicar un año particularmente desafiante en términos de seguridad vial.
- Es notable la marcada reducción de accidentes en 2020, con 22,000 incidentes registrados. Este descenso podría estar relacionado con las restricciones de movilidad impuestas durante la pandemia de COVID-19.
- A partir de 2019, observamos una disminución progresiva en el número de accidentes. El descenso continúa en 2020 y 2021. Entender los factores de este descenso puede llevarnos a plantear estrategias que permitan la constante mejoría de la seguridad vial.
- Aunque el 2021 muestra una disminución adicional, con 18,000 accidentes, es vital investigar si las medidas implementadas en respuesta a la pandemia han influido en la disminución continua de la accidentalidad

```python
df['FECHA_OCURRENCIA_ACC'] = pd.to_datetime(df['FECHA_OCURRENCIA_ACC'])
df['año'] = df['FECHA_OCURRENCIA_ACC'].dt.year
accidentes_por_año = df['ANO_OCURRENCIA_ACC'].value_counts().sort_index()

plt.plot(accidentes_por_año.index, accidentes_por_año.values, marker='o')
plt.title('Accidentes por Año')
plt.xlabel('Año')
plt.ylabel('Número de Accidentes')
plt.show()
```

### Accidentes por hora del dia

- Al seccionar cierta parte del conjunto de datos, podemos abrir algunas posibilidades para analizar, de manera que podemos ser muy especificos, y ver en que horas se producen la mayor cantidad de accidentes; de manera que podemos ver que entre las 12 y las 3 de la tarde, se producen una mayor cantidad de accidentes, ¿significa esto que es gracias a la hora?..., exactamente no, esto puede ser debido a muchas razones, la gente quiere volver a su trabajo, hay un mayor flujo de vehiculos publicos, y nos volvemos menos conscientes del riesgo constante que hay en la carretera. Teniendo esto en cuenta, podemos implementar soluciones como un mejor control vehicular por parte de las entidades policiacas, una llamada a la consciencia de la gente para ser mas precavidos, si bien en todo momento, más en algunas horas.

```python
df['FECHA_HORA_ACC'] = pd.to_datetime(df['FECHA_HORA_ACC'])

df_subset = df[(df['FECHA_HORA_ACC'].dt.hour >= 0) & (df['FECHA_HORA_ACC'].dt.hour < 24)]

accidentes_por_hora = df_subset['FECHA_HORA_ACC'].dt.hour.value_counts().sort_index()

plt.plot(accidentes_por_hora.index, accidentes_por_hora.values, marker='o')
plt.title('Accidentes por Hora del Día (Subconjunto)')
plt.xlabel('Hora del Día')
plt.ylabel('Número de Accidentes')
plt.show()
```

### Accidentes por mes

- El entender porque en ciertos meses hay un mayor indice de accidentalidad no puede ayudar a pensar correctamente en las soluciones, en este caso podemos desmentir un paradigma que algunas personas podrían tener, y es que al estar en temporada vacacional, la gente viaja mucho, por lo que habrán más accidentes, y en base a la grafica podemos ver que es lo contrario, ya que si bien hay bastante accidentes, vemos que cuando se va llegando a junio disminuyen, y lo mismo en diciembre, por lo que estas temporadas no son un factor altamente de riesgo, y de esta forma podemos llevar a cabo medidas de control  y diferentes proyectos que hagan que la accidentalidad disminuya.

```python
df['mes'] = df['FECHA_OCURRENCIA_ACC'].dt.month

accidentes_por_mes = df['mes'].value_counts().sort_index()

plt.plot(accidentes_por_mes.index, accidentes_por_mes.values, marker='o')
plt.title('Accidentes por Mes')
plt.xlabel('Mes')
plt.ylabel('Número de Accidentes')
plt.show()
```
